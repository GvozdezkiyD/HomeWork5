## Задача 1: Нахождение экстремумов функции

Найдем экстремумы функции:

$$
f(x, y, z) = 2x^3 + 2xy + 2xz + y^2 + z^2 + 2y - 8
$$

### Шаг 1: Найдем частные производные

#### 1. Частная производная по \( x \):

Для вычисления частной производной по переменной \( x \), используем формулу:

$$
f_x = \frac{\partial f}{\partial x} = \frac{\partial}{\partial x} \left( 2x^3 + 2xy + 2xz + y^2 + z^2 + 2y - 8 \right)
$$

Разберём каждый член:

- \( 2x^3 \) по \( x \) даёт \( 6x^2 \),
- \( 2xy \) по \( x \) даёт \( 2y \),
- \( 2xz \) по \( x \) даёт \( 2z \),
- \( y^2 \), \( z^2 \), \( 2y \), \( -8 \) не содержат переменную \( x \), их производные по \( x \) равны нулю.

Итак, частная производная по \( x \):

$$
f_x = 6x^2 + 2y + 2z
$$

#### 2. Частная производная по \( y \):

Теперь найдём частную производную по переменной \( y \):

$$
f_y = \frac{\partial f}{\partial y} = \frac{\partial}{\partial y} \left( 2x^3 + 2xy + 2xz + y^2 + z^2 + 2y - 8 \right)
$$

Разберём каждый член:

- \( 2x^3 \) не зависит от \( y \), его производная равна 0,
- \( 2xy \) по \( y \) даёт \( 2x \),
- \( 2xz \) не зависит от \( y \), его производная равна 0,
- \( y^2 \) по \( y \) даёт \( 2y \),
- \( z^2 \), \( -8 \) не зависят от \( y \), их производные равны 0,
- \( 2y \) по \( y \) даёт \( 2 \).

Итак, частная производная по \( y \):

$$
f_y = 2x + 2y + 2
$$

#### 3. Частная производная по \( z \):

Найдем частную производную по переменной \( z \):

$$
f_z = \frac{\partial f}{\partial z} = \frac{\partial}{\partial z} \left( 2x^3 + 2xy + 2xz + y^2 + z^2 + 2y - 8 \right)
$$

Разберём каждый член:

- \( 2x^3 \), \( 2xy \) не зависят от \( z \), их производные равны 0,
- \( 2xz \) по \( z \) даёт \( 2x \),
- \( y^2 \), \( 2y \), \( -8 \) не зависят от \( z \), их производные равны 0,
- \( z^2 \) по \( z \) даёт \( 2z \).

Итак, частная производная по \( z \):

$$
f_z = 2x + 2z
$$

### Шаг 2: Составление системы уравнений

Для нахождения критических точек приравняем частные производные к нулю:

f_x = 6x^2 + 2y + 2z = 0 \\

f_z = 2x + 2z = 0
\end{cases}
$$

Решая эту систему уравнений, мы получаем критические точки.

### Шаг 3: Анализ второго порядка (матрица Гессе)

Для того, чтобы классифицировать найденные экстремумы (минимум, максимум или седловая точка), нужно исследовать матрицу Гессе \( H \), которая состоит из вторых частных производных функции.

Запишем вторые частные производные:

#### Вторая частная производная по \( x \):

$$
f_{xx} = \frac{\partial^2 f}{\partial x^2} = 12x
$$

#### Вторая частная производная по \( y \):

$$
f_{yy} = \frac{\partial^2 f}{\partial y^2} = 2
$$

#### Вторая частная производная по \( z \):

$$
f_{zz} = \frac{\partial^2 f}{\partial z^2} = 2
$$

#### Смешанные вторые производные:

$$
f_{xy} = f_{yx} = 2
$$

$$
f_{xz} = f_{zx} = 2
$$

$$
f_{yz} = f_{zy} = 0
$$

Теперь составим матрицу Гессе:

$$
H = \begin{pmatrix}
f_{xx} & f_{xy} & f_{xz} \\
f_{yx} & f_{yy} & f_{yz} \\
f_{zx} & f_{zy} & f_{zz}
\end{pmatrix}
= \begin{pmatrix}
12x & 2 & 2 \\
2 & 2 & 0 \\
2 & 0 & 2
\end{pmatrix}
$$

Для классификации экстремума используем детерминант матрицы Гессе.

1. Если \( \det(H) > 0 \) и \( f_{xx} > 0 \), то точка является **локальным минимумом**.
2. Если \( \det(H) > 0 \) и \( f_{xx} < 0 \), то точка является **локальным максимумом**.
3. Если \( \det(H) < 0 \), то точка является **седловой**.
4. Если \( \det(H) = 0 \), то анализ не даёт однозначного результата.

Теперь вычислим детерминант матрицы Гессе для каждой из найденных точек критического числа.

### Шаг 4: Анализ второго порядка

Для классификации найденных точек экстремума необходимо исследовать матрицу Гессе \( H \), которая состоит из вторых частных производных функции \( f \).

#### 1. Вычисление вторых частных производных:

Матрица Гессе для функции \( f(x, y, z) \) определяется как:

$$
H =
\begin{pmatrix}
\frac{\partial^2 f}{\partial x^2} & \frac{\partial^2 f}{\partial x \partial y} & \frac{\partial^2 f}{\partial x \partial z} \\
\frac{\partial^2 f}{\partial y \partial x} & \frac{\partial^2 f}{\partial y^2} & \frac{\partial^2 f}{\partial y \partial z} \\
\frac{\partial^2 f}{\partial z \partial x} & \frac{\partial^2 f}{\partial z \partial y} & \frac{\partial^2 f}{\partial z^2}
\end{pmatrix}
$$

Вычислим вторые производные:

- \( \frac{\partial^2 f}{\partial x^2} = \frac{\partial}{\partial x} (6x^2 + 2y + 2z) = 12x \)
- \( \frac{\partial^2 f}{\partial y^2} = \frac{\partial}{\partial y} (2x + 2y + 2) = 2 \)
- \( \frac{\partial^2 f}{\partial z^2} = \frac{\partial}{\partial z} (2x + 2z) = 2 \)
- \( \frac{\partial^2 f}{\partial x \partial y} = \frac{\partial}{\partial y} (6x^2 + 2y + 2z) = 2 \)
- \( \frac{\partial^2 f}{\partial x \partial z} = \frac{\partial}{\partial z} (6x^2 + 2y + 2z) = 2 \)
- \( \frac{\partial^2 f}{\partial y \partial z} = \frac{\partial}{\partial z} (2x + 2y + 2) = 0 \)

Таким образом, матрица Гессе:

$$
H =
\begin{pmatrix}
12x & 2 & 2 \\
2 & 2 & 0 \\
2 & 0 & 2
\end{pmatrix}
$$

#### 2. Оценка знака определителя матрицы Гессе:

Чтобы определить тип критической точки, вычислим детерминант матрицы Гессе \( \det(H) \) для каждой найденной точки.

##### Для точки \( (1, -2, -1) \):

Подставим \( x = 1 \) в матрицу Гессе:

$$
H_1 =
\begin{pmatrix}
12 \cdot 1 & 2 & 2 \\
2 & 2 & 0 \\
2 & 0 & 2
\end{pmatrix}
=
\begin{pmatrix}
12 & 2 & 2 \\
2 & 2 & 0 \\
2 & 0 & 2
\end{pmatrix}
$$

Вычислим детерминант \( H_1 \):

$$
\det(H_1) = 12(2 \cdot 2 - 0 \cdot 0) - 2(2 \cdot 2 - 0 \cdot 2) + 2(2 \cdot 0 - 2 \cdot 2)
$$

$$
\det(H_1) = 12(4) - 2(4) + 2(-4) = 48 - 8 - 8 = 32
$$

Поскольку \( \det(H_1) > 0 \) и \( f_{xx} = 12x = 12 > 0 \), то точка \( (1, -2, -1) \) является **локальным минимумом**.

##### Для точки \( \left( -\frac{1}{3}, -\frac{2}{3}, \frac{1}{3} \right) \):

Подставим \( x = -\frac{1}{3} \) в матрицу Гессе:

$$
H_2 =
\begin{pmatrix}
12 \cdot \left( -\frac{1}{3} \right) & 2 & 2 \\
2 & 2 & 0 \\
2 & 0 & 2
\end{pmatrix}
=
\begin{pmatrix}
-4 & 2 & 2 \\
2 & 2 & 0 \\
2 & 0 & 2
\end{pmatrix}
$$

Вычислим детерминант \( H_2 \):

$$
\det(H_2) = -4(2 \cdot 2 - 0 \cdot 0) - 2(2 \cdot 2 - 0 \cdot 2) + 2(2 \cdot 0 - 2 \cdot 2)
$$

$$
\det(H_2) = -4(4) - 2(4) + 2(-4) = -16 - 8 - 8 = -32
$$

Поскольку \( \det(H_2) < 0 \), то точка \( \left( -\frac{1}{3}, -\frac{2}{3}, \frac{1}{3} \right) \) является **седловой точкой**.

### Шаг 5: Запись результата и выводов

Мы нашли две критические точки для функции \( f(x, y, z) = 2x^3 + 2xy + 2xz + y^2 + z^2 + 2y - 8 \):

1. **Точка \( (1, -2, -1) \):**
   - **Тип экстремума:** Локальный минимум

### Шаг 1: Метод Лагранжа

Для нахождения экстремумов функции с ограничением используем метод множителей Лагранжа. Вводим множитель Лагранжа \( \lambda \) и составляем Лагранжиан:

$$
\mathcal{L}(x, y, \lambda) = f(x, y) - \lambda \cdot g(x, y)
$$

Подставляем выражения для \( f(x, y) \) и \( g(x, y) \):

$$
\mathcal{L}(x, y, \lambda) = 4x + 8y - \lambda (y^2 - 2xy + 5)
$$

### Шаг 2: Частные производные

Теперь находим частные производные Лагранжиана по переменным \( x \), \( y \) и \( \lambda \), и приравниваем их к нулю.

#### Частная производная по \( x \):

$$
\frac{\partial \mathcal{L}}{\partial x} = 4 - \lambda \cdot (-2y) = 0
$$
Или:
$$
4 + 2\lambda y = 0 \quad \Rightarrow \quad \lambda = -\frac{2}{y}
$$

#### Частная производная по \( y \):

$$
\frac{\partial \mathcal{L}}{\partial y} = 8 - \lambda \cdot (2y - 2x) = 0
$$
Или:
$$
8 - \lambda(2y - 2x) = 0 \quad \Rightarrow \quad \lambda = \frac{8}{2y - 2x} = \frac{4}{y - x}
$$

#### Частная производная по \( \lambda \):

$$
\frac{\partial \mathcal{L}}{\partial \lambda} = -(y^2 - 2xy + 5) = 0
$$
Или:
$$
y^2 - 2xy + 5 = 0
$$

### Шаг 3: Решение системы уравнений

Теперь мы имеем систему уравнений:

1. \( \lambda = -\frac{2}{y} \)
2. \( \lambda = \frac{4}{y - x} \)
3. \( y^2 - 2xy + 5 = 0 \)

Приравняем выражения для \( \lambda \):

$$
-\frac{2}{y} = \frac{4}{y - x}
$$

Решим это уравнение относительно \( x \):

$$
-2(y - x) = 4y
$$
$$
-2y + 2x = 4y
$$
$$
2x = 6y \quad \Rightarrow \quad x = 3y
$$

Теперь подставим \( x = 3y \) в третье уравнение:

$$
y^2 - 2(3y)y + 5 = 0
$$
$$
y^2 - 6y^2 + 5 = 0
$$
$$
-5y^2 + 5 = 0
$$
$$
y^2 = 1 \quad \Rightarrow \quad y = \pm 1
$$

### Шаг 4: Нахождение \( x \)

Теперь, зная, что \( y = 1 \) или \( y = -1 \), подставим эти значения в \( x = 3y \):

- Для \( y = 1 \): \( x = 3(1) = 3 \)
- Для \( y = -1 \): \( x = 3(-1) = -3 \)

### Шаг 5: Проверка решений

Таким образом, мы получаем два возможных решения:

1. \( (x, y) = (3, 1) \)
2. \( (x, y) = (-3, -1) \)

### Шаг 6: Проверка типа экстремума с помощью второго порядка

Для проверки типа экстремума (минимум или максимум) нужно использовать вторые производные

```python
import sympy as sp

# Определяем переменные
x, y, lambda_ = sp.symbols('x y lambda')

# Функция и ограничение
f = 4*x + 8*y
g = y**2 - 2*x*y + 5

# Лагранжиан
lagrange = f - lambda_ * g

# Частные производные по x, y и lambda
f_x = sp.diff(lagrange, x)
f_y = sp.diff(lagrange, y)
f_lambda = sp.diff(lagrange, lambda_)

# Решаем систему уравнений
solutions = sp.solve([f_x, f_y, f_lambda], (x, y, lambda_))

# Выводим решение
solutions
```

**Задача 3** (30 баллов)

Вам предложены данные с информацией о успеваемости студентов:

Независимые переменные:
* Hours Studied: Общее количество часов, потраченных на учебу каждым студентом.
* Previous Scores: Баллы, полученные студентами на предыдущих экзаменах.
* Sleep Hours: Среднее количество часов сна студента в сутки.
* Sample Question Papers Practiced: Количество пробных экзаменационных работ, с которыми студент занимался.
  
Целевая переменная:
* Performance Index: Показатель общей успеваемости каждого студента. Индекс успеваемости отражает академическую успеваемость студента и округляется до ближайшего целого числа. Индекс варьируется от 10 до 100, при этом более высокие значения свидетельствуют о более высокой успеваемости.

**Решите задачу линейной регрессии, реализовав градиентный спуск самостоятельно, не используя готовое решение из какой-либо библиотеки.**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression

# Исходные данные (Hours Studied, Previous Scores, Sleep Hours, Sample Question Papers Practiced, Performance Index)
data = np.array([
    [1, 2, 8, 3, 70],
    [2, 3, 7, 4, 75],
    [1.5, 2.5, 6.5, 3.5, 72],
    [3, 4, 6, 5, 80],
    [2.5, 3.5, 7, 4, 78],
    [4, 5, 5, 6, 85],
    [3.5, 4.5, 6, 5, 82],
    [2, 3, 8, 4, 74],
    [2.2, 2.8, 7.5, 4.2, 76],
    [3, 4, 6.5, 5, 79]
])

# Разделим данные на независимые переменные (X) и целевую переменную (y)
X = data[:, :-1]  # Все столбцы, кроме последнего
y = data[:, -1]   # Последний столбец (Performance Index)

# Стандартизируем данные (нормализуем их)
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Добавим столбец единиц (для свободного члена θ0)
X = np.c_[np.ones(X.shape[0]), X]

# Функция ошибки (среднеквадратичная ошибка)
def compute_cost(X, y, theta):
    m = len(y)
    predictions = X.dot(theta)
    cost = (1 / (2 * m)) * np.sum((predictions - y) ** 2)
    return cost

# Градиентный спуск
def gradient_descent(X, y, theta, learning_rate, iterations):
    m = len(y)
    cost_history = []

    for i in range(iterations):
        gradient = (1 / m) * X.T.dot(X.dot(theta) - y)
        theta -= learning_rate * gradient
        cost_history.append(compute_cost(X, y, theta))

    return theta, cost_history

# Инициализация переменных
theta_initial = np.zeros(X.shape[1])
learning_rate = 0.1
iterations = 1000

# Запуск градиентного спуска
theta_optimal, cost_history = gradient_descent(X, y, theta_initial, learning_rate, iterations)

# Результаты
print("Оптимальные коэффициенты модели (θ):", theta_optimal)
print("Минимальная стоимость (ошибка):", cost_history[-1])

# Визуализация уменьшения стоимости
plt.plot(range(iterations), cost_history)
plt.xlabel('Итерации')
plt.ylabel('Стоимость (MSE)')
plt.title('Уменьшение стоимости в процессе градиентного спуска')
plt.show()

# Предсказания с использованием оптимальных коэффициентов
y_pred = X.dot(theta_optimal)

# Оценка точности модели (среднеквадратичная ошибка)
mse = mean_squared_error(y, y_pred)
print("Среднеквадратичная ошибка (MSE):", mse)

# Сравнение с библиотечным решением
model = LinearRegression()
model.fit(X[:, 1:], y)  # Используем только стандартизированные данные
print("Коэффициенты модели sklearn:", model.coef_)
print("Свободный член (θ0) sklearn:", model.intercept_)
print("Среднеквадратичная ошибка (MSE) sklearn:", mean_squared_error(y, model.predict(X[:, 1:])))
```
